```python
import pandas as pd
import numpy as np
import os
```

```python
df=pd.read_csv("../data/stores_sales_forecasting.csv", encoding="latin1")
```

```python
df.head()
df.info()
```

```python
df.columns=[df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('-','_')]
```

```python
df.info()
```

```python
df.head()
```

```python
df.columns
```

The CSV was read with a multi-level header, so columns were tuples. I flattened and standardized them before applying transformations.

```python
df.columns = [
    col[0] if isinstance(col, tuple) else col
    for col in df.columns
]
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['ship_date'] = pd.to_datetime(df['ship_date'])
df.info()
df.columns
```

```python
df = df.drop_duplicates(subset=['order_id', 'product_id'])
```

```python
# Postal code often missing - not needed for forecasting
df['postal_code'] = df['postal_code'].fillna('UNKNOWN')

# Validate critical fields
df = df.dropna(subset=['order_date', 'sales', 'region', 'category'])
```

Creating Time Features

```python
df['order_week'] = df['order_date'].dt.to_period('W').apply(lambda r: r.start_time)
df['year'] = df['order_date'].dt.year
df['month'] = df['order_date'].dt.month
df['week_of_year'] = df['order_date'].dt.isocalendar().week
df.head()
```
